```
import json
def crop_image(image_path, ground_truth_path, save_img_dir, save_img_labels):
  label_dict = dict()
  img_list = os.listdir(image_path)
  for j in range(len(img_list)):
    image = img_list[j]
    b = os.path.join(image_path, image)
    a =os.path.join(ground_truth_path, 'gt_' + image.replace("jpg", "txt"))
    labels = []
    boxes = []
    with open(a, encoding = "utf-8") as f:
      file = f.readlines()
    
      for i in range(len(file)):
        file[i] = re.sub("\n", "", file[i])
    
      for line in file:
        if '\ufeff' in line:
          line = line.replace("\ufeff", "")
        info = line.split(",")

        coor = np.asarray(info[0:8], dtype = np.int32)
        label =  info[8]
        boxes.append(coor)
        labels.append(label)



    for i in range(len(boxes)):
      box = boxes[i]
      img = cv2.imread(b)
      img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
      pts = np.array([[box[0],box[1]],[box[2],box[3]],[box[4],box[5]],[box[6],box[7]]])

      rect = cv2.boundingRect(pts)
      x,y,w,h = rect
      cropped = img[y:y+h, x:x+w].copy()
  
    ## (2) make mask
      pts = pts - pts.min(axis=0)

      mask = np.zeros(cropped.shape[:2], np.uint8)
      cv2.drawContours(mask, [pts], -1, (255), -1, cv2.LINE_AA)

    ## (3) do bit-op
      dst = cv2.bitwise_and(cropped, cropped, mask=mask)
      # dst_string = np.array2string(dst)
      
      # save_img_name = "image_" +str(j)+ "_" + str(i) + ".png"
      # save_img_name = "image_" +str(j)+ "_" + str(i) + ".txt"
      # with open(os.path.join(save_img_dir, save_img_name), mode = "w") as f:
      #     f.write(dst_string)

      
      save_img_name = "image_" +str(j)+ "_" + str(i) + ".npy"
     
      np.save(os.path.join(save_img_dir, save_img_name), dst)


      # cv2.imwrite(os.path.join(save_img_dir, save_img_name), dst)

      label_dict[save_img_name] = labels[i]
  
  
  with open(save_img_labels + ".json", 'w') as json_file:
    json.dump(label_dict, json_file)
  print(f"Number of labels: {len(label_dict)}")
  print(f"Number of images: {len(os.listdir(save_img_dir))}")
  print("Finished")
```

```
image_path = 'icdar2015/ch4_test_images'
ground_truth_path =  'icdar2015/ch4_test_gt'
save_img_dir = "icdar2015/gray_test"
save_img_labels = "gray_test_labels"
crop_image(image_path, ground_truth_path, save_img_dir, save_img_labels)
```

```
image_path = 'icdar2015/ch4_training_images'
ground_truth_path =  'icdar2015/ch4_training_gt'
save_img_dir = "icdar2015/gray_train"
save_img_labels = "gray_train_labels"
crop_image(image_path, ground_truth_path, save_img_dir, save_img_labels)
```
